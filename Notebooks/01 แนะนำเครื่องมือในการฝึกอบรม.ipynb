# **แนะนำเครื่องมือในการฝึกอบรม**

### **Client Tool ที่ใช้อยู่เรียกว่า Azure Data Studio มีความสามารถในการสร้าง Notebook ที่ผู้ฝึกอบรมเปิดใช้งานอยู่ขณะนี้ ซึ่งมีความสามารถในการแสดง Cell ใน 2 รูปแบบ**

- **Text Cell ก็คือส่วนของข้อความที่ผู้ฝึกอบรมกำลังอ่านอยู่ในขณะนี้**
- **Code Cell คือ Cell ที่ปรากฎด้านใต้ Cell นี้ ท่านสามารถสังรัน แก้ไข Code สั่งรัน ท่านผู้ฝึกอบรมจะเห็นผลลัพธ์ต่อท้าย Cell ทันที**

```sql
PRINT 'This is Code Cell';

--ทดลองเปลี่ยนคำสั่ง PRINT เป็นคำสั่ง SELECT แล้วลองอีกครั้ง
```

# **แนะนำฐานข้อมูลที่ใช้ในการฝึกอบรม**

### ฐานข้อมูล TSQL <span style="color: var(--vscode-foreground);">เป็นเรื่องราวทางธุรกิจเกี่ยวกับ สินค้า, ลูกค้า, การสั่งสินค้า, ผู้จัดจำหน่าย และพนักงาน</span>



ผู้ฝึกอบรมจะสังเกตเห็นว่ามีวงเล็บท้ายชื่อตาราง ดังนี้ Sales, Production และ HR เป็นชื่อ schema (ขอบเขตด้านความปลอดภัยของ Objects ในหลักสูตรนี้เราจะไม่ลงรายละเอียดของ schema) การอ้างถึง objects จะต้องระบุ schema ด้วย ดัง Code ด้านล่าง

```sql
USE TSQL;

SELECT * FROM HR.Employees;
```
